# Hilbert space filling curve

As mentioned here I am using Hilbert curves which are a fractal mapping technique to preserve grouping. As other written work on the internet discussing Hilbert curves I follow the tradition to include [this](https://xkcd.com/195/) relevant xkcd.

Generating Hilbert and in general fractal curves is best described using the L-system also called Lindenmayer system. This is a compact way to describe iterative graphics and for a Hilbert curve the rule is given by:

Starting from $L$ replace $L -> +RF-LFL-FR+$ and $R -> -LF+RFR+FL-$ with $90^0$ rotations. It took me a while to get how this works so I will summarize it here. In the Lindenmayer system we start with an Axiom and we apply the rewrite rules iteratively. The system contains an alphabet which holds the characters to be rewritten and constants which do not change. Here $F$ means move forward and $+,-$ means anticlockwise and clockwise rotations. Also the Alphabet is ignored when we are drawing. Finally, we need an origin point which in our case is $(0, 0)$ pointing the the direction of $(1, 0)$. The full system is shown below:

Alphabet: $L, R$

Constants: $F + −$

Axiom : $L$

Production rules:

$L -> +RF-LFL-FR+$

$R -> -LF+RFR+FL-$

All the movements are done for a unit of $1$. To illustrate I will use the [HilbertSpaceFillingCurve.jl](https://github.com/jonathanBieler/HilbertSpaceFillingCurve.jl) package.

First we need to create a function which will perform the drawing and store the coordinated returned from the package.

```julia
using HilbertSpaceFillingCurve

function hilb_plot(s::Int64)
    x::Array{Int64,1}, y::Array{Int64,1} = Int[], Int[]
    for d = 0:1:s^2-1
        hilb = hilbert(d, 2, 16)
        push!(x, hilb[1]) 
        push!(y, hilb[2])
    end
    
    return x, y
end
```

Here I will draw the first four Hilbert curves.

```julia
using Plots
x_1, y_1 = hilb_plot(2)
plot_1 = plot(x_1, y_1, title = "Iteration #1")

x_2, y_2 = hilb_plot(4)
plot_2 = plot(x_2, y_2, title = "Iteration #2")

x_3, y_3 = hilb_plot(8)
plot_3 = plot(x_3, y_3, title = "Iteration #3")

x_4, y_4 = hilb_plot(16)
plot_4 = plot(x_4, y_4, title = "Iteration #4")

plot(plot_1, plot_2, plot_3, plot_4, legend=false, fmt=:svg, width=2)
```

As a sanity check we can do the first curve. The string we get is $+RF-LFL-FR+$. Again all rotations are $90^0$ and we start with anticlockwise, move one step, rotate clockwise, move one step, rotate clockwise and move one step. This traces the first figure. Finally, [this tool](https://onlinemathtools.com/l-system-generator) allows you to explore the rewrite rules.

# Fitting a 2D sinusoid

Before fitting an actual image we can try a 2D sinusoid from $0$ to $\pi$ using 4 pixels per side. For this we need the Colors and Images packages.

```julia
using Colors, Images
```

We then simply write a function to generate a 2D sinusoid.

```julia
function sin_2d(;s = 4, range_min = 1.0, range_max = 1.0)
    p = ones(s, s)
    r_x = collect(range(0.0, length=s, stop=range_min*π))
    r_y = collect(range(range_max*π, length=s, stop=0.0))

    for i in collect(1:s)
        for j in collect(1:s)
            p[j, i] = sin(r_x[i]) + sin(r_y[j])
        end
    end
    p = p / 2.
    
    return p
end

dim8 = 8
p8 = sin_2d(s=dim8)
img8 = Gray.(p8)
```

Pretty pixelated, but hey it's only 64 pixels. Next we need to store the pancaked version of the image where the read path is defined by a Hilbert curve.

```julia
function hilbertize!(x::Array{Int64,1}, y::Array{Int64,1}, 
                     himg::Array{Float64,1}, p::Array{Float64,2}; s = 4)
    for d = 0:1:s^2-1
        hilb = hilbert(d, 2, 16)
        push!(x, hilb[1]) 
        push!(y, hilb[2])
        himg[d+1] = p[hilb[1]+1, hilb[2]+1]
    end
end
```

And then call the function to perform the fractal traverse.

```julia
himg8 = zeros(dim8^2)
x8, y8 = Int[], Int[]
hilbertize!(x8, y8, himg8, p8, s=dim8);
```

Next we can plot the coordinates of the Hilbert curve and see the path which was used for the traverse and the extracted pixel grayscale value based on that path.

```julia
p8_hilb = plot(x8, y8, title="64 px")
p8_data = plot(himg8)
plot(p8_hilb, p8_data, legend=false, fmt=:svg, layout=(2, 1), width=2)
```

We can now compare between different sampling density and the idea behind the Hilbert curve is that we should preserve the shape. For this we can use our $8\times 8$ pixel image and compare it with a say $16\times 16$ one.

```julia
dim16 = 16
p16 = sin_2d(s=dim16)
img16 = Gray.(p16)
```

As expected more detail is shown in the image.

```julia
himg16 = zeros(dim16^2)
x16, y16 = Int[], Int[]
hilbertize!(x16, y16, himg16, p16, s=dim16);
```

And we can compare the results with the $8 \times 8$ pixel image.

```julia
p16_hilb = plot(x16, y16, title="256 px")
p16_data = plot(himg16)
plot(p8_hilb, p16_hilb, p8_data, p16_data, legend=false, fmt=:svg, layout=(2, 2), width=2)
```

The result shows us that the shape is preserved and the increase of sampling points leads to more information or noise depending on how successful we are in making this into sound.

From inspecting the figures I think the best approach to "see" what is going on in the picture is to look for the gradients. If we can distinguish between the high and low values we can maybe built intuition about the shape in the picture.

# Comparing pancake grouping

At this point it might the worthwhile to consider what would happen if we do not use a Hilbert curve to generate the extraction path, but instead flatten (pancake) the image.

```julia
flatten_p8 = p8[:]

plot(plot(himg8, title="Hilbert 64px"),
     plot(flatten_p8, title="Flatten 64px"),
     legend=false, fmt=:svg, width=2)
```

On the left we can see the generated path by a Hilbert curve and on the right is the flattened path. By itself we cannot say much about the comparison, however, when we change the resolution maybe we will see something interesting.

```julia
flatten_p16 = p16[:]

plot(plot(himg16, title="Hilbert 256px"),
     plot(flatten_p16, title="Flatten 256bpx"),
     legend=false, fmt=:svg, width=2)
```

It is clear that the change in resolution produced a completely different shape in the version using a flattened path, however, it was preserved in the one using a Hilbert path. With confidence we can now proceed to generate sounds.

# Sounds in Julia

To play the function we generated we can create sounds in Julia using the WAV library. The sounds consist of a time axis measured in miliseconds and the value axis which is the sound value.

```julia
using WAV
```

The Nyquist frequency is half of the theoretical sampling rate of a discrete digital system. Given that human audible hearing is generally between $20 \text{Hz}$ and $20 \text{kHz}$ the Nyquist frequency for our case is $40 \text{kHz}$. However, because real signals are not perfect a higher frequency than the Nyquist is recommended. For our case I will use the ubiquitous sampling rate of $44.1 \text{kHz}$. So first of all how to we create a wave file in Julia and play it. It is remarkably easy! The sine wave we are going to plot is in Equation 1.

$$
\begin{equation}
y(x, t) = sin(2 \pi f t)
\end{equation}
$$

```julia
f = 44100
time_data = 0:1:f*1
sound_data = sin.(2 * π * 440 * time_data / f)

wavplay(sound_data, f)
wavwrite(sound_data, "440Hz.wav", Fs=f)
```
